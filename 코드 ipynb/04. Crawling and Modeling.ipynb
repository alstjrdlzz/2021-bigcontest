```python
driver.get("https://www.koreabaseball.com/Schedule/Schedule.aspx#none")

collect_data = []


for team in range(2, 12):
    team_ = driver.find_element_by_xpath('//*[@id="contents"]/ul/li[%s]/a'%team).text
    driver.find_element_by_xpath('//*[@id="contents"]/ul/li[%s]/a'%team).click()
    for j in range(1, 13):
        MM = driver.find_element_by_xpath('//*[@id="ddlMonth"]/option[%s]'%j).text
        driver.find_element_by_xpath('//*[@id="ddlMonth"]/option[%s]'%j).click()


        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        for num in range(1, 160):
            if len(soup.select('#tblSchedule > tbody > tr:nth-of-type(%s) > td.day'%num)) == 1:
                day = soup.select('#tblSchedule > tbody > tr:nth-of-type(%s) > td.day'%num)

            time = soup.select('#tblSchedule > tbody > tr:nth-of-type(%s) > td.time'%num)
            play1 = soup.select('#tblSchedule > tbody > tr:nth-of-type(%s) > td.play > span:nth-of-type(1)'%num)
            em = soup.select('#tblSchedule > tbody > tr:nth-of-type(%s) > td.play > em'%num)
            play2 = soup.select('#tblSchedule > tbody > tr:nth-of-type(%s) > td.play > span:nth-of-type(2)'%num)

            if (len(time) == 0) & (len(play1) == 0) & (len(em) == 0) & (len(play2) == 0):
                pass
            else:
                collect_data.append([team_, day[0].text, time[0].text, play1[0].text, em[0].text, play2[0].text])
             
```

```python
경기일정 = pd.DataFrame(collect_data, columns=['T_ID','mmdd','time','play1','em','play2'])
경기일정['YYMMDD'] = '2021-'+경기일정.mmdd.str.split('(').str[0].str.replace('.','-')
경기일정 = 경기일정[경기일정.em!='vs']


for i in 경기일정.T_ID.unique():
    경기일정.loc[경기일정[경기일정.T_ID==i].sort_values(['YYMMDD','time']).index, '경기수'] = range(1, len(경기일정[경기일정.T_ID==i])+1)
    
경기일정['YYMMDDTT'] = 경기일정.YYMMDD.astype('str')+'_'+경기일정.time

경기일정.loc[경기일정[경기일정.duplicated(['YYMMDD','T_ID'], keep='last')].index , 'double'] = 1
경기일정.loc[경기일정[경기일정.duplicated(['YYMMDD','T_ID'], keep='first')].index , 'double'] = 2

경기일정_ = 경기일정.fillna(0)

경기일정_ = team_rename(경기일정_, 'T_ID', create_col = None)
경기일정_ = team_rename(경기일정_, 'play1', create_col = None)
경기일정_ = team_rename(경기일정_, 'play2', create_col = None)



경기일정_.to_excel(os.getcwd()+'/data/경기일정.xlsx')
```

```python
# short term 데이터 생성

def 짧은데이터(col, dataset):
    LAG_ = pd.DataFrame()
    PRED = pd.DataFrame()
    for i in dataset.PCODE.unique():
        a = dataset[dataset.PCODE==i].sort_values('경기수')
        a__ = a.copy().reset_index(drop=True)
        
        for j in range(1,6):
            col_ = ['%s_lag_%s'%(col, j)]
        
            a_ = a[['%s'%col]].shift(-j).reset_index(drop=True).rename(columns=dict(zip(['%s'%col], col_)))
            a__ = pd.concat([a__,a_],axis=1).dropna()
            
        PRED = pd.concat([PRED, a__[-21:]],axis=0).dropna()
        
        for j in [28]:
        
            col_ = ['target_%s'%j]

            a_ = a[['%s'%col]].shift(-j).reset_index(drop=True).rename(columns=dict(zip(['%s'%col], col_)))
            a__ = pd.concat([a__,a_],axis=1).dropna()
            
        LAG_ = pd.concat([LAG_, a__],axis=0).dropna()
    LAG_ = LAG_.reset_index(drop=True)
    
    train = pd.DataFrame()
    test = pd.DataFrame()
    
    for i in LAG_.PCODE.unique():
        A1 = LAG_[LAG_.PCODE==i].reset_index(drop=True)

        train = pd.concat([train, A1[A1.index<len(A1) - 30]])
        test = pd.concat([test, A1[A1.index>=len(A1) - 30]])

    x_train_, x_test_, y_train_, y_test_ = train[train.columns[2:-1]], test[test.columns[2:-1]],  train[['PCODE','target_28']], test[['PCODE','target_28']]

    return x_train_, x_test_, y_train_, y_test_, PRED.reset_index(drop=True)


# grid serach

def 나는_그리드(x_train, x_test, y_train, y_test):
    
    svm_reg_param_grid = [
    {'kernel': ['linear'], 'C':[10., 30., 100., 300., 1000., 3000., 10000., 30000.0]},
    {'kernel': ['rbf'], 'C': [1.0, 3.0, 10., 30., 100., 300., 1000.0],
                            'gamma':[0.01, 0.03, 0.1, 0.3, 1.0, 3.0]},
    ]
    #SVR
    svm_reg = SVR()
    svm_reg_grid_search = GridSearchCV(svm_reg, svm_reg_param_grid, cv =3, scoring = "neg_mean_squared_error", n_jobs= -1 )
    svm_reg_grid_search.fit(x_train, y_train)
    #svm_reg_짱 = svm_reg_grid_search.best_params_

    
    
    rf_param_grid = {
        'bootstrap': [True],
        'max_depth': [8, 16, 32, 64],
        'max_features': [1, 2, 3],
        'min_samples_leaf': [1, 3, 5, ],
        'min_samples_split': [3, 5, 7],
        'n_estimators': [300, 500, 1000],
        'random_state' : [seed]
    }
    rf = RandomForestRegressor()
    # Instantiate the grid search model
    rf_grid_search = GridSearchCV(estimator = rf, param_grid = rf_param_grid, cv = 3, n_jobs = -1, verbose = 2)
    rf_grid_search.fit(x_train, y_train)
    #rf_짱 = rf_grid_search.best_params_
    
    
    
    GBR_search_grid={'loss':['quantile'],'n_estimators':[500,1000],'learning_rate':[.001,0.01,.1],'max_depth':[1,2,4],'subsample':[.5,.75,1],'random_state':[seed]}
    GBR=GradientBoostingRegressor()
    GBR_search=GridSearchCV(estimator=GBR, param_grid=GBR_search_grid, scoring='neg_mean_squared_error', n_jobs=1)
    GBR_search.fit(x_train, y_train)
    #GBR_짱 = GBR_search.best_params_
    
    
    xgb_parameters = {'nthread':[4], #when use hyperthread, xgboost may become slower
                  'objective':['reg:linear'],
                  'learning_rate': [0.01, 0.1], #so called `eta` value
                  'max_depth': [5, 6, 7],
                  'min_child_weight': [4],
                  'silent': [1],
                  'subsample': [0.1, 0.3, 0.5, 0.7, 0.9],
                  'colsample_bytree': [0.5, 0.7, 0.9],
                  'n_estimators': [500, 700, 900],
                  'reg_lamda': [1, 2, 3],
                    "random_state":[seed]}
    xgb = xgboost.XGBRegressor()
    xgb_grid = GridSearchCV(xgb,xgb_parameters,cv = 3,n_jobs = 5,verbose=True)
    xgb_grid.fit(x_train, y_train)
    #xgb_짱 = xgb_grid.best_params_
    
    LGBM_param_grid = {
        'n_estimators': [1500, 2000, 2500],
        'colsample_bytree': [0.7, 0.8],
        'max_depth': [15, 20, 25],
        'num_leaves': [100, 200],
        'reg_alpha': [ 0.3, 0.5, 0.01],
        'reg_lambda': [0.1, 1, 3, 5],
        'min_split_gain': [0.2, 0.3],
        'subsample': [0.7, 0.8],
        'subsample_freq': [1, 5, 10, 20],
        'learning_rate':[0.21],
        "random_state":[seed]
    }
    
    lgbm = LGBMRegressor()
    # Instantiate the grid search model
    lgbm_grid_search = GridSearchCV(estimator = lgbm, param_grid = LGBM_param_grid, cv = 3, n_jobs = -1, verbose = 2)
    lgbm_grid_search.fit(x_train, y_train)
    #lgbm_grid_짱 = lgbm_grid_search.best_params_
    
    return svm_reg_grid_search, rf_grid_search, GBR_search, xgb_grid, lgbm_grid_search



```
